# .apply()

```python
s = pd.Series(['Male', 'male', 'Female', 'f', 'female', 'm', 'Male', 'Female'])
s
```

```python
s.unique()
```

# We want to make this data consistant and accurete

```python
def clean(string):
  return string.upper()[0]
```

```python
clean(string="Male")
```

```python
clean(string= "f")
```

```python
s_new = s.apply(clean)
```

```python
s_new
```

```python
car = pd.Series(['toyota', 'bmw', 'supra', 'ferrari', 'thar', 'maruti', 'fortuner'])
```

```python
def series_upper(string):
  return string.upper()
```

```python
new_car = car.apply(series_upper) # defining fuction
```

```python
new_car
```

```python
s.apply(lambda x: x.upper()) # using lambda function
```

# Vertically concate Series to get DataFrame

```python
s1 = pd.Series(['Rohit', 'Simran','Jane'], name= 'Student')
s2 = pd.Series(['M', 'F', 'F'], name= 'Gender')
df = pd.concat([s1,s2], axis= 'columns')
```
