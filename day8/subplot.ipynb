```python
import numpy as np
x = np.linspace(-5,5, 100)
```

```python
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
ax = axes[0,0]
ax.plot(x,sinx, label='sinx')
ax.legend()
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

ax = axes[0,1]
ax.plot(x,cosx, label="cosx")
ax.legend()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
ax =axes[1,0]
ax.plot(x,tanx, label="tanx")
ax.legend()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

# for [0,0]
ax = axes[0,0]
ax.plot(x,sinx, label='sinx')
ax.legend()

# for [0,1]
ax = axes[0,1]
ax.plot(x,cosx, label='cosx')
ax.legend()

# for [1,0]
ax = axes[1,0]
ax.plot(x,tanx, label='tanx')
ax.legend()
```

# Annoting Barplot

```python
gender = ["Male", "Female"]
frequency = [187, 87]
fig, axes = plt.subplots()
axes.bar(x=gender, height=frequency)
axes.bar_label(axes.containers[0])
fig.show()
```
