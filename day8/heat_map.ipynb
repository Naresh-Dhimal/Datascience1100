# HeatMap
- The relationship between two variables is shown by correlation and we can visualize it using HeatMap

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
```

```python
df.head(3)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

# Annoting heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

* The correlation bewteen petal_length and petal_width is 0.96. It is positive correlation. It means when the value of peal_length increases the value of petal_width also increase. You can visualixe this using scatterplot.

```python
sns.scatterplot(x='petal_length', y ='petal_width', data=df)
```
