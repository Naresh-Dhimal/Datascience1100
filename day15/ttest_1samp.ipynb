# t-test

```python
import numpy as np
```

```python
mu = 50 # dairy milk claims
```

```python
np.random.seed(0)
sample_weight = np.random.randint(low=46, high=55, size=25)
```

```python
sample_weight
```

```python
H0 = "There is no difference in weight"
H1 = "There is siginificant in weight "
alpha = 0.05
```

```python
# t-test: we want to compare the mean of reference and sample
from scipy import stats
result = stats.ttest_1samp(sample_weight, popmean=mu)
result
```

```python
if result.pvalue < alpha:
  print(H1)
else:
  print(H0)
```

```python
if result.pvalue < alpha:
  print("we reject H0")
else:
  print("we fail to reject H0")
```
