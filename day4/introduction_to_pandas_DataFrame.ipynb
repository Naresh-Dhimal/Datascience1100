# Creating a DataFrame using list of list

```python
data = [['e01', 'kshitiz'],['e02', 'shyam'],['e03', 'Rita']]
```

```python
df = pd.DataFrame(data)
df
```

# how to add column name in list of list

```python
# first approach
data = [['e01', 'kshitiz'],['e02', 'shyam'],['e03', 'Rita']]
df = pd.DataFrame(data, columns = ['emp_id', 'emp_name'])
df
```

```python
# another approach
data = [['e01', 'kshitiz'],['e02', 'shyam'],['e03', 'Rita']]
df = pd.DataFrame(data)
df.columns = ['emp_id', 'emp_name']
```

```python
df
```

# to update columns variable

```python
data  = {
    "country" : ['Nepal', 'India', 'China', 'Japan', 'USA', 'France'],
    "capital" : ['Kathmandu', 'New Delhi', 'Beijing', 'Tokyo', 'Washington DC','Paris' ]
}
df = pd.DataFrame(data)
df.rename({'capital': 'capital_city'}, axis= 'columns', inplace=True) # update the df variable
df
```

# Deleting a column in DataFrame

# first approach

```python
dfs = pd.DataFrame({'c1':['v1','v2'], 'col 2':['v22','v23']})
dfs
```

```python
del dfs['col 2']
```

```python
dfs
```

# another approach

```python
dfs = pd.DataFrame({'c1':['v1','v2'], 'col 2':['v22','v23']})
dfs
```

```python
dfs.drop(columns=['col 2'], inplace=True)
```

```python
dfs
```
