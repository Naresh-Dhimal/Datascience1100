# Introductiton matplotlib

```python
# initializing matplotlib
import matplotlib.pyplot as plt
# initializing seaborn as sns
import seaborn as sns
```

# Types of data analysis:
* Univariate: one variables
* Bivariate: two variables
* Multi-variate: three or more variables

# Univariate
* line chart

```python
month = ['jan', 'feb', 'march','april','may', 'june']
growth = [3, 3.6, 2, 2.3, -2, 5]

plt.plot(month, growth, color= 'Purple')
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.show()
```

```python
month = ['jan', 'feb', 'march','april','may', 'june']
growth = [3, 3.6, 2, 2.3, -2, 5]

plt.plot(month, growth, 'o-', color= 'Purple')
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.show()
```

```python
plt.plot(month, growth, 'o--', color= 'Purple')
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.show()
```

```python
plt.plot(month, growth, 'o--', color= 'Purple')
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.grid()
plt.show()
```

```python
month = ['jan', 'feb', 'march','april','may', 'june']
growth = [3, 3.6, 2, 4, 5, 5]
expected = [2,2.5,4,4.5,5,6]

plt.plot(month, growth, 'o--', color= 'Purple')
plt.plot(month, expected, 'o-', color= 'blue')
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.show()
```

```python
# labeling
month = ['jan', 'feb', 'march','april','may', 'june']
actual_growth = [3, 3.6, 2, 4, 5, 5]
expected_growth = [2,2.5,4,4.5,5,6]

plt.plot(month, growth, 'o--', color= 'Purple', label = "actual_growth")
plt.plot(month, expected, 'o-', color= 'blue', label = "expected_growth")
plt.xlabel("Month")
plt.ylabel("Growth(%)")
plt.legend() # plots the label in plt.plot()
plt.show()
```

* Bar Graph

```python
gender = ['Female','Male']
count = [4, 8]
plt.bar(x=gender,height=count)
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
# using seaborn
sns.barplot(x=gender, y=count)
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```
