```python
import pandas as pd
```

```python
df = pd.read_excel("/content/employee-dept-table.xlsx")
df
```

```python
employee = pd.read_excel("/content/employee-dept-table.xlsx", sheet_name = 'employee')
department = pd.read_excel("/content/employee-dept-table.xlsx", sheet_name ='dept')
```

```python
employee
```

```python
department
```

# Inner join - intersection join(by default)

```python
pd.merge(employee, department, on = "DepartmentID", how='inner')
```

```python
pd.merge(employee,department)
```

# Full outer join - union join


```python
pd.merge(employee, department , on= "DepartmentID", how ="outer")
```

# Left Join

```python
pd.merge(employee, department, on= 'DepartmentID', how="left")
```

```python
# NaN means missing values; the departmentId 106 has no department name and location so value is missing
```

```python
pd.merge(department, employee, on= 'DepartmentID', how="left")
```

# Right join

```python
pd.merge(employee, department, on= 'DepartmentID', how="right")
```

```python
pd.merge(department, employee, on= 'DepartmentID', how="right")
```
