```python
# to find mean of engine cylinder of "Subaru" with "Transmission Type = AUTOMATIC"
round(df[(df['Make']== 'Subaru') & (df['Transmission Type']=='AUTOMATIC')]['Engine Cylinders'].mean(),2)
```

```python
# here, mean of engine cylinder of "Subaru" with "Transmission Type = AUTOMATIC" is  4.62
```

```python
df.loc[1:3,:] # slicing rows 1 to 3 here, loc is used so location of data 1 to 3 is present and for column all is present
```

```python
# .loc[]
df.loc[1:4, 'Make':'Engine HP'] # slicing rows 1 to 4 here, loc is used so location of data 1 to 4 is present and for column from "Make": "Engine HP"
```

```python
# .iloc[]
df.iloc[1:4, 1:4] # slicing rows 1 to 4 here, iloc is used so location of data 1 to 3 is present and for column from 1 to 3
```
