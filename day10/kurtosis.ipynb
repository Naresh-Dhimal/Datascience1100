# Kurtosis
Kurtosis refers to the degree of presence of outliers in the distribution . Kurtosis is a statistical measure, whether the datat is heavy-tailed or light-tailed in a normal.

There are 3 types of kurtosis:
1. Mesokurtic
2. Platykurtic
3. Leptokurtic

# MesoKurtic:
* Normal Distribution: Mean = $\mu$; varience = $\sigma^2$
* Medium outlier
* kurtosis = 3

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed(0)
data = np.random.normal(loc=5, scale = 1, size = 1500)
kurt = kurtosis(data, fisher=False) # fisher = False means pearson
plt.title(f"Kurtosis = {round(kurt, 2)}")
sns.histplot(data, kde=True)
plt.show()
```

# Platykurtic:
* Uniform distribution
* No outliers
* Kurtosis < 3

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis

np.random.seed(0)
data = np.random.uniform(low=3, high=6, size =15000)
kurt = kurtosis(data, fisher=False) # fisher = False means pearson
plt.title(f"Kurtosis = {round(kurt, 2)}")
sns.histplot(data, bins=3)
plt.show()
```

# Leptokurtic:
* Laplace distribution
* high outliers
* kurtosis > 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

np.random.seed(0)
data = np.random.laplace(loc=5, scale=1, size=1500)
kurtosis = kurtosis(data, fisher=False)
plt.title(f"Kurtosis = {round(kurtosis, 2)}")
sns.histplot(data, kde=True)
plt.show()
```
