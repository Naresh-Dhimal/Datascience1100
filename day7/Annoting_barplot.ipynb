#  Annoting barplot

```python
# using plotly
fruits = ['Mango', 'Orange', 'Pineapple', 'Kiwi']
count = [5, 8, 9, 15]
import plotly.express as px
px.bar(x=fruits, y=count, text=count)
```

```python
# change height and width
fig = px.bar(x=fruits, y=count, text=count)
fig.update_layout(height=400, width=600, xaxis_title ='Fruits', yaxis_title ='Count')
```

# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
fruits = ['Mango', 'Orange', 'Pineapple', 'Kiwi']
count = [5, 8, 9, 15]
plt.figure(figsize=(6,2))
plt.pie(count, labels=fruits)
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.pie(count, labels=fruits, autopct= "%.2f%%")
plt.show()
```

# Plotly

```python
fig = px.pie(names=fruits, values=count)
fig.update_layout(width = 500, height=500 )
```

# Stacked Barplot

```python
month= ['jan', 'Feb','Mar','Apr']
male_present = [120, 112, 80, 140]
female_present = [110, 80, 50, 130]
```

```python
plt.bar(month, male_present, label ='Male_present')
plt.bar(month, female_present, label ='Female_present')
plt.legend()
plt.show()
```

# Groupby barplot using pandas

```python
# Creating data Frame
import pandas as pd
df = pd.DataFrame({
    "Month": month,
    "male_present": male_present,
    "female_present": female_present
})

# plotting graph
df.plot(x="Month", y=['male_present', "female_present"], kind= "bar")
plt.show()
```

# Boxplot and Histogram and Kdeplot

```python
sns.boxplot(df['sepal_width'])
plt.show()
```

# Histogram and Kdeplot
* Used for continuous data<br>
`Continuous data is data that can take any value`<br>
`Frequency  distribution of continuous data is called histogram`

```python
# Sepal_length is continuous data
```

```python
plt.hist(df['sepal_length'])
```

```python
# to control bin
plt.hist(df['sepal_length'], bins=5)
```

```python
# using seaborn
sns.histplot(df['sepal_width'])
```

```python
sns.histplot(df['sepal_width'], kde=True)
```

```python
sns.kdeplot(df['sepal_width'], fill=True)
```
